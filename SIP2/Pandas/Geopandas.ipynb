```python
bounds = buildings.total_bounds #bbox dla całego zbioru
bounds
```

```python
buildings.bounds.head(10) #bbox dla każdego obiektu
```

```python
import shapely.geometry
polygon_buildings=shapely.geometry.box(*bounds)
type(polygon_buildings)
#polygon_buildings
```

```python
bbox=(364955.9307,360321.9223,369037.3682,362657.6811) #bbox wybranego fragmentu miasta
type(bbox)
```

```python
polygon_ww = shapely.geometry.box(*bbox) #tworzenie obiektu o geometrii polygon z bbox
polygon_ww
```

```python
sel=buildings.intersects(polygon_ww) #wybór budynków, które są w poligonie
sel
```

```python
sel.value_counts()
```

```python
ww=buildings[sel] #utworzenie GeoDataFrame zawierającej wybrane budynki
```

```python
ww.set_crs(2180, allow_override=True)
```

```python
fig, ax = plt.subplots(figsize=(15, 15))
ww.plot(column='LKOND',
        figsize=(9, 15),
        ax=ax,
        legend=True,
        cmap='Set2',
        categorical=True,
        )
ax.set_title('Liczba kondygnacji budynków', fontdict={'fontsize': '25'})
```

```python
ww.area
```

```python
ww.reset_index()
```

```python
ww.area.iloc[6]
```

```python
ww['area_km2'] = ww.area / 1000**2
```

```python
ww.head(3)
```

```python
centroids=ww.geometry.centroid
type(centroids)
```

```python
centroids_df=gpd.GeoDataFrame(geometry=centroids)
```

```python
centroids_df
```

```python
fig, ax = plt.subplots(figsize=(15, 15))
centroids_df.plot(
        figsize=(9, 15),
        ax=ax,
        markersize=4,
        categorical=True,
        )
```

```python
centroids_df.to_crs(epsg=4326).to_file('/content/drive/MyDrive/Colab Notebooks/build_cent.geojson',driver='GeoJSON')
#wizualizacja na https://geojson.io/
```
