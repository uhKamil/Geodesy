##Podstawy Pandas

Pandas to to nowoczesna, potężna i bogata w funkcje biblioteka przeznaczona do analizy danych w Pythonie. Jest to dojrzały framework do analizy danych, który jest szeroko stosowany w różnych dziedzinach nauki. W związku z tym istnieje dokumentacja i wiele dobrych przykładów, które mogą pomóc w realizacji zadań związanych z analizą danych.

W pandas dane są zwykle przechowywane w strukturze danych zwanej DataFrame, która wygląda jak typowa tabela z wierszami i kolumnami (+ indeksy i nazwy kolumn), gdzie kolumny mogą zawierać dane różnych typów. 

Jedną z najbardziej użytecznych funkcji pandas jest możliwość bezpośredniego odczytu danych z wielu różnych formatów danych. Na przykład pandas obsługuje odczytywanie i zapisywanie danych z/do: CSV, JSON, HTML,MSExcel, HDF5, Format Pythona Pickle, SQL (Postgresql, MySQL, Oracle, itp.)

## Tworzenie Data Frame

```python
lst = ['Analizy przestrzenne', 
       'Automatyzacja analiz geoprzestrzennych', 
       'Wprowadzenie do Python', 
       'Uczenie maszynowe', 
       'Nauka gotowania']
```

```python
df=pd.DataFrame(lst)
df
```

```python
df=pd.DataFrame(lst,columns=['Nazwa_kursu'])
df
```

```python
df=pd.DataFrame(lst,index=['a','b','c','d','e'],columns=['Nazwa_kursu'])
df
```

```python
df=pd.DataFrame({'Nr_kursu':[1,2,3],
                 'Nazwa_kursu':['Analizy przestrzenne',"Automatyzacja analiz geoprzestrzennych",'Wprowadzenie do Python']})
df
```

## Selekcja danych

`.loc[]` pozwala na na wybór wierszy po nazwach.

`.iloc[]` pozwala na wybór wierszy po indeksie.

```python
df = pd.DataFrame(
    {
     'Nr_kursu': {'w1': '1', 'w2':'2', 'w3':'3'}, 
     'Nazwa_kursu': {'w1':'Analizy przestrzenne', 'w2':'Automatyzacja analiz geoprzestrzennych','w3':'Wprowadzenie do Python'},
    })

df
```

```python
df.loc[['w1']] #pierwszy wiersz
```

```python
df.loc[['w1'],['Nazwa_kursu']]
```

```python
df.iloc[0:2]
```

```python
df.loc[df.Nazwa_kursu=='Wprowadzenie do Python']
```

## Selekcja cd.

Jednym ze sposobów wybierania wierszy z DataFrame jest podanie listy wartości, które są używane do znajdowania pasujących wierszy w określonej kolumnie DataFrame za pomocą funkcji isin().

```python
selection=['Marszowice','Rędzin','Pilczyce'] #selekcja wierszy zawierających wartości 'Marszowice','Rędzin','Pilczyce' w kolumnie 'NAZWA"
dem.loc[dem["NAZWA"].isin(selection)]
```

```python
dem.loc[(dem['W0_2']>10) & (dem['LUD_NA_KM2']<100)]
```
